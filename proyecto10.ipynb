### ¿Qué tipo de establecimiento es habitualmente una cadena?

```python
chain_is_true=rest_data[rest_data['chain']==True]
business_type_chain=chain_is_true.groupby('object_type')[['object_type']].count()
business_type_chain=business_type_chain.rename(columns={'object_type':'count_business_type_chain'})
business_type_chain.reset_index(inplace=True)
business_type_chain
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8, 8))
plt.pie(business_type_chain['count_business_type_chain'], labels=business_type['object_type'],autopct='%1.1f%%', startangle=120, colors=sns.color_palette("GnBu", len(business_type_chain)))
plt.title('Proporciones de los tipos de establecimientos que pertenecen a una cadena')
plt.axis('equal')  
plt.show()
```

Los establecimientos que pertenecen a una cadena revela que la mayoría de ellos son del tipo "Restaurant", seguido por "Fast Food" y "Bakery".Estos tres tipos de establecimientos constituyen una parte significativa del total, lo que indica que las cadenas suelen centrarse en estos sectores de la industria alimentaria.

```python
rest_data = rest_data.dropna(subset=['chain'])
rest_data['chain'] = pd.to_numeric(rest_data['chain'], errors='coerce')
pivot = pd.pivot_table(rest_data, index='object_type', values='chain', aggfunc='mean')
pivot['chain'] = pivot['chain'] * 100
pivot['Non-chain Percentage'] = 100 - pivot['chain']
pivot = pivot.sort_values(by='chain', ascending=False)
pivot = pivot.rename(columns={'chain': 'Chain Percentage'})
```

```python
fig, ax = plt.subplots(figsize=(13, 9))
bar_width = 0.35
index = np.arange(len(pivot.index))
chains = ax.bar(index, pivot['Chain Percentage'], bar_width, color='b', label='Cadena')
non_chains = ax.bar(index + bar_width, pivot['Non-chain Percentage'], bar_width, color='cyan', label='No Cadena')
ax.set_xlabel('Tipo de Establecimiento')
ax.set_ylabel('Porcentaje')
ax.set_title('Comparación de establecimientos de cadena y no cadena por tipo')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(pivot.index)
ax.yaxis.set_major_formatter(mtick.PercentFormatter(100))
ax.legend();
```

Un gráfico excelente Jesus que nos permite visualizar de la mejor forma la relación entre cadenas y no cadenas segun los tipos de establecimiento! Sigamos!

Las panaderías son 100% cadenas se puede  esperar que sean de propiedad familiar. Los cafés y la comida rápida están detrás con alrededor del 60%, los restaurantes y bares son menos del 40% de cadenas.


### ¿Qué caracteriza a las cadenas: muchos establecimientos con un pequeño número de asientos o unos pocos establecimientos con un montón de asientos?

```python
rest_data[rest_data['chain']==True]['number']
```

```python
sns.set_style("white")
plt.figure(figsize=(13,9))
sns.histplot(data=rest_data[rest_data['chain']==True], x='number', color='blue')
plt.title('Distribución de asientos en cadenas de restaurantes')
plt.xlabel('Numero de asientos')
plt.ylabel('Número de Establecimientos')
sns.despine();
```

<div class="alert alert-block alert-success">

Aquí lo hemos hecho perfecto Jesus! Hemos utilizado el dataframe filtrado con las cadenas y hemos aplicado un histograma sobre los asientos que nos permite ver la distribución que buscabamos. |

```python
chains_df = rest_data.query('chain == True').groupby('object_name').agg({'id':'count', 'number':'mean'}).reset_index()

chains_df.columns = ['name', 'number_of_restaurants', 'average_number_of_seats']

chains_df
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12, 8))
sns.scatterplot(y='number_of_restaurants', x='average_number_of_seats', data=chains_df, color='blue', s=20, edgecolor='w', alpha=0.7)
plt.title('Distribución de asientos en cadenas de restaurantes', fontsize=16)
plt.xlabel('Numero de Asientos', fontsize=14)
plt.ylabel('Número de Establecimientos', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
sns.despine(trim=True)
plt.show()
```

La mayoría de los establecimientos de cadena tienen entre 1 y 50 asientos. La mitad de ellos tienen al menos 25 .Esto tiene sentido porque, como se vio antes, es más probable que las panaderías, cafeterías y restaurantes de comida rápida sean sucursales de cadenas y que tengan menos asientos.

### Coloca los datos de los nombres de las calles de la columna address en una columna separada.

```python
rest_data['street'] = rest_data['address'].str.extract(r'\s(.+)')
rest_data.head()
```

```python
num_rest_by_streets = rest_data.groupby('street')[['object_type']].count().reset_index()
num_rest_by_streets = num_rest_by_streets.rename(columns= {'object_type': 'count_rest'})
num_rest_by_streets.head()
```


### Traza un gráfico de las diez mejores calles por número de restaurantes.

```python
top_10_streets = num_rest_by_streets.sort_values(by= 'count_rest', ascending= False).head(10)
```

```python
top_10_streets_sorted = top_10_streets.sort_values(by='count_rest', ascending=False)
labels = top_10_streets_sorted['street']
sizes = top_10_streets_sorted['count_rest']
plt.figure(figsize=(12, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Blues', len(labels)))
plt.title('Top 10 de Calles con más Establecimientos', fontsize=16)
plt.axis('equal')
plt.show()
```
